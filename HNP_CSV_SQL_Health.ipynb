```python
health_df = pd.read_csv(r'./World bank data/HNP_StatsData.csv')

health_dfCopy = health_df.copy()
health_dfCopy = health_dfCopy.rename(columns={'Country Name': 'CountryName'})
health_dfCopy  = health_dfCopy .drop(health_dfCopy [(health_dfCopy .CountryName != "Canada") &
                                                                      (health_dfCopy.CountryName != "United States") & 
                                                                      (health_dfCopy.CountryName != "Mexico") &
                                                                     (health_dfCopy.CountryName != "Mauritius") &
                                                                      (health_dfCopy.CountryName != "Liberia") &
                                                                      (health_dfCopy.CountryName != "Dem. Rep. Congo") &
                                                                      (health_dfCopy.CountryName != "Vietnam") &
                                                                      (health_dfCopy.CountryName != "Nepal") &
                                                                      (health_dfCopy.CountryName != "Bangladesh") 
                                                                      ].index)

health_dfCopy = health_dfCopy.rename(columns={'Indicator Name': 'Indicator_Name'})
health_dfCopy  = health_dfCopy .drop(health_dfCopy [(health_dfCopy.Indicator_Name != "Domestic general government health expenditure per capita (current US$)") &
                                                                      (health_dfCopy.Indicator_Name != "Hospital beds (per 1,000 people)") & 
                                                                      (health_dfCopy.Indicator_Name != "Immunization, HepB3 (% of one-year-old children)") &
                                                                     (health_dfCopy.Indicator_Name != "Physicians (per 1,000 people)") &
                                                                      (health_dfCopy.Indicator_Name != "Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)") &
                                                                      (health_dfCopy.Indicator_Name != "Number of people who are undernourished") &
                                                                      (health_dfCopy.Indicator_Name != "Prevalence of overweight (% of adults)") &
                                                                      (health_dfCopy.Indicator_Name != "Prevalence of overweight (% of children under 5)") &
                                                                      (health_dfCopy.Indicator_Name != "Completeness of death registration with cause-of-death information (%)") &
                                                                      (health_dfCopy.Indicator_Name != "Low-birthweight babies (% of births)") &
                                                                      (health_dfCopy.Indicator_Name != "Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)") &
                                                                      (health_dfCopy.Indicator_Name != "Adults (ages 15+) and children (0-14 years) living with HIV") &
                                                                      (health_dfCopy.Indicator_Name != "Prevalence of current tobacco use (% of adults)") &
                                                                      (health_dfCopy.Indicator_Name != "Out-of-pocket expenditure (% of current health expenditure)") &
                                                                      (health_dfCopy.Indicator_Name != "Mortality from CVD, cancer, diabetes or CRD between exact ages 30 and 70 (%)") &
                                                                      (health_dfCopy.Indicator_Name != "Nurses and midwives (per 1,000 people)")
                                                                      ].index)
health_dfCopy = health_dfCopy.drop(['1960', '1961', '1962', '1963', '1964',
       '1965', '1966', '1967', '1968', '1969', '1970', '1971', '1972', '1973',
       '1974', '1975', '1976', '1977', '1978', '1979', '1980', '1981', '1982',
       '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991',
       '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000',
       '2001', '2002', '2003', '2004', 'Country Code', 'Indicator Code', 'Unnamed: 66', '2021'], axis=1)

countrys = ["Canada", "United States", "Mexico", "Mauritius", "Liberia", "Dem. Rep. Congo", "Vietnam", "Nepal", "Bangladesh"]
df = pd.DataFrame()
for x in range(len(countrys)):
                     
                     country_thing = health_dfCopy.loc[(health_dfCopy['CountryName'] == countrys[x])]
                     country_thing=country_thing.drop(['CountryName'],axis=1)
                     #print(canada)
                     #health_dfCopy =pd.pivot_table(health_dfCopy,index=["Indicator_Name"])
                     country_thing.set_index('Indicator_Name',inplace=True)
                     health_dfCopy1 = country_thing[["2005","2006","2007","2008","2009","2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"]]
                     health_dfCopy1 = health_dfCopy1.T
                     health_dfCopy1 = health_dfCopy1.reset_index()
                     health_dfCopy1 = health_dfCopy1.rename(columns={ "Domestic general government health expenditure per capita (current US$)": "Domestic_government_health_expenditure_per_capita(current US$)",
                                                               "Hospital beds (per 1,000 people)": "Hospital_beds(per_1,000_people)",
                                                               "Immunization, HepB3 (% of one-year-old children)": "Immunization_HepB3(%_of_one-year-old_children)",
                                                               "Physicians (per 1,000 people)": "Physicians(per_1,000_people)",
                                                               "Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)": "Deaths_communicable_diseases,maternal_prenatal&nutrition(%)",
                                                               "Number of people who are undernourished":"Number_of_people_who_are_undernourished",
                                                               "Prevalence of overweight (% of adults)":"Prevalence_of_overweight(%_of_children_under_5)",
                                                               "Prevalence of overweight (% of children under 5)":"Prevalence_of_overweight(%_of_adults)",
                                                               "Completeness of death registration with cause-of-death information (%)":"Completeness_of_death_registration_with_cause-of-death_info(%)",
                                                               "Low-birthweight babies (% of births)":"Low-birthweight_babies(%_of_births)",
                                                               "Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)":"Total_alcohol_consumption_per_capita(L_of_pure_alcohol,15+)",
                                                               "Adults (ages 15+) and children (0-14 years) living with HIV":"Adults(ages_15+)_and_children(0-14_years)_living_with_HIV",
                                                               "Prevalence of current tobacco use (% of adults)":"Prevalence_of_current_tobacco_use(%_of_adults)",
                                                               "Out-of-pocket expenditure (% of current health expenditure)":"Out-of-pocket_expenditure(%_of _current_health_expenditure)",
                                                               "Mortality from CVD, cancer, diabetes or CRD between exact ages 30 and 70 (%)":"Mortality_from_CVD,cancer,diabetes_or_CRD_between_30&70(%)",
                                                               "Nurses and midwives (per 1,000 people)":"Nurses_and_midwives(per_1,000_people)",
                                                               "index":"year"})
                     country = [countrys[x]] * 16
                     health_dfCopy1['Country']=country
                     
                     if df.empty:
                            df = health_dfCopy1
                     else:
                            df = df.append(health_dfCopy1, ignore_index=True)

print(df)

```
